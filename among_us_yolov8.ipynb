```python
from ultralytics import YOLO

# Load the YOLOv8n model
model = YOLO("yolov8m.pt")

# Train the model with essential parameters
model.train(
    data="/content/amongusdata/data.yaml",  # Path to the dataset YAML file
    epochs=50,                   # Number of epochs
    imgsz=640,                   # Image size
    batch=8                      # Batch size
)
```

```python
# Save the trained model
model_path = "/content/amongusdata/yolov8_model.pt"
model.save(model_path)
print(f"Model saved to {model_path}")

```

```python
# Load the trained model
model = YOLO(model_path)

# Evaluate the model on the validation dataset
# Evaluate the model on the validation dataset
metrics = model.val(data="/content/amongusdata/data.yaml")

# Print the evaluation metrics
print(metrics)

# Optionally, you can evaluate on a separate test dataset if you have one
# metrics = model.val(data="/path/to/your/test_data.yaml")
# print(metrics)
```
